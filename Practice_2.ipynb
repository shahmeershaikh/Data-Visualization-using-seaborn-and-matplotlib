```python
plt.figure(figsize = (5, 4))
sns.countplot(data = df, x = 'gender', hue = 'gender', palette = {'female': 'pink', 'male': 'blue'})
plt.show()
```

```python
df['gender'].value_counts()
```

```python
gender = df['gender']. value_counts().index.tolist()
gender_color = ['pink', 'lightblue']
gender_freq = df ['gender']. value_counts().tolist()
print(gender, gender_freq)
```

```python
plt.figure(figsize = (4, 4))
plt.pie(gender_freq, labels = gender, colors = gender_color, autopct = '%1.2f%%')
plt.show()
```

```python
plt.figure(figsize = (4, 4))
plt.pie(gender_freq, labels = gender, colors = gender_color, autopct = '%1.2f%%',
       startangle = 90, wedgeprops = {"width": 0.7
                                    })
plt.show()
```
