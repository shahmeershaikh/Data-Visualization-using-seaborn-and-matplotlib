```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
plt.figure(figsize = (5, 4))
sns.histplot(data = df, x = 'math_score', color = 'g')
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.histplot(data = df, x = 'writing_score', color = 'g', fill = False)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.histplot(data = df, x = 'math_score', color = 'skyblue', binwidth = 10, alpha = 0.5)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.histplot(data = df, x = 'reading_score', color = 'g', kde = True, alpha = 0.3)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.kdeplot(data = df, x = 'math_score', fill = True, color = 'yellow')
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.scatterplot(data = df, x = 'reading_score', y = 'math_score', color = 'orange')
plt.show
```

```python
plt.figure(figsize = (5, 4))
sns.scatterplot(data = df, x = 'math_score', y = 'reading_score', hue = 'gender',
               marker = 'o', s = 50, alpha = 0.6, palette = {'female':'pink', 'male':'blue'})
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.scatterplot(data = df, x = 'writing_score', y = 'reading_score', hue = 'gender',
               marker = 'o', s = 50, alpha = 0.6, palette = {'female':'pink', 'male': 'blue'})
plt.grid(True)
plt.show()
```

```python
sns.regplot (x='math_score', y='reading_score',data=df, scatter_kws={'color':'skyblue'}, line_kws={'color':'red'})
```
