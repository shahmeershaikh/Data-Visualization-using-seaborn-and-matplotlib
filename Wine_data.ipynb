```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winequality.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
new_df = df.iloc[:, 0:11]
new_df.head()
```

```python
coco = new_df.corr()
coco
```

```python
plt.figure(figsize = (9, 7))
sns.heatmap(coco, square = True, cmap = 'bwr', linewidth = 0.5, vmin =-1,vmax =1, annot = True)
plt.xticks(rotation = 65)
plt.show()
```

```python
tri_matrix = np.triu(coco)
print(tri_matrix)
```

```python
plt.figure(figsize = (5, 5))
sns.heatmap(coco, square = True, cmap = 'managua', linewidth = 0.7, vmin = -1, vmax = 1,
           mask = tri_matrix, annot = True, fmt = '.1f')
plt.xticks(rotation = 65)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.scatterplot(data = df, x = 'fixed acidity', y = 'alcohol', color = 'orange')
plt.show()
```

```python
plt.figure(figsize = (5, 5))
sns.kdeplot(data = df, x = 'fixed acidity', y = 'alcohol', color = 'blue', alpha = 0.3)
plt.show()
```

```python
plt.figure(figsize = (6, 5))
sns.scatterplot(data = df, x = 'fixed acidity', y = 'alcohol', color = 'orange')
sns.kdeplot(data = df, x = 'fixed acidity', y = 'alcohol', color = 'blue', alpha = 0.3)
plt.show()
```

```python
sns.jointplot(data = df, x = 'fixed acidity', y = 'alcohol', kind = 'reg')
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.kdeplot(data = df, x = 'alcohol', y = 'sulphates', hue = 'quality',
           palette = 'RdPu', bw_adjust = 1, levels = 5)
plt.show()
```
