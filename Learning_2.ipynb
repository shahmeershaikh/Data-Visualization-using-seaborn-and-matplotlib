```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
sns.catplot(data = df, x = 'gender', y = 'math_score', hue = 'gender',
            kind = 'box', palette = {'male':'skyblue', 'female': 'pink'})
plt.show()
```

```python
plt.figure(figsize = (5, 5))
sns.violinplot(data = df, x = 'gender', y = 'math_score', hue = 'gender',
              palette = {'male':'skyblue', 'female': 'pink'})
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.pointplot(data = df, x = 'gender', y = 'math_score', linestyle = '--')
plt.title('Mean and 95% Confident Interval of Math Score by Gender')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.pointplot(data = df, x = 'ethnicity', y = 'math_score', linestyle = '--')
plt.grid(True)
plt.show()

print('Group E has the highest mean and Group A has the lowest mean')
```

```python
plt.figure(figsize = (5, 4))
sns.pointplot(data = df, x = 'ethnicity', y = 'math_score', linestyle = '--', hue = 'gender',
             palette = {'male':'skyblue', 'female': 'pink'})
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.kdeplot(data = df, x = 'math_score', hue = 'gender', common_norm = 'False', fill = 'True',
           palette = {'male':'skyblue', 'female':'pink'}, lw = 2)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.pointplot(data=df, x="ethnicity", y="math_score", errorbar=("pi", 100),
              capsize=.2, color="hotpink", linestyle="none", marker="D",
              mfc='blue', mec='skyblue', markersize=10)
plt.title('Math Score Gap by Ethnicity')
plt.grid()
plt.show()
```
