```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 50, 500)
y = np.log(x)
```

```python
plt.figure(figsize = (5, 4))
plt.gca().set_facecolor('skyblue')
plt.plot(x, y, color = 'green', linestyle = '-', linewidth = 4)
plt.grid(True)
plt.show()
```

```python
months = ['August', 'September', 'October', 'November', 'December']
income = [5000, 3000, 4000, 1500, 5200]
```

```python
plt.figure(figsize = (5, 4))
sns.lineplot(x = months, y = income, color = 'blue', marker = 'X', markersize = 8, linestyle = '--')
plt.grid(True)
plt.show()
```
